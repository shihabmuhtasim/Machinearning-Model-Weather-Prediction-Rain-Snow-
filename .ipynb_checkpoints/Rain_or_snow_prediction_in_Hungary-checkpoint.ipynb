**7. Standardization scaling**

```python
import copy

unscaled_data = copy.deepcopy(weather_data)
```

```python
from sklearn.preprocessing import StandardScaler

# Separate the target variable
a = weather_data.drop('Precip Type', axis=1)

# Initialize the StandardScaler
scaler = StandardScaler()


scaled_features = scaler.fit_transform(a)

# Update the original weather_data  with the scaled features
weather_data[a.columns] = scaled_features

```

```python
weather_data.head()
```

**PCA LEFT TO DO**

**MODEL 1: DECESION TREE**

```python
weather_data.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score
```

```python
imputer = SimpleImputer(strategy='mean')  # Replace missing values with the mean
X = weather_data.drop('Precip Type', axis=1)
y = weather_data['Precip Type']
X_imputed = imputer.fit_transform(X)
```

```python
# 2. Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.3, random_state=128)
```

```python
# 3. Train the DecisionTreeClassifier
dtClassifier = DecisionTreeClassifier(max_leaf_nodes=15, random_state=0)
dtClassifier.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_des = dtClassifier.predict(X_test)
```

```python
y_pred_des[:100]
```

```python
y_test[:100]
```

```python
# Calculate accuracy
accuracy_des = accuracy_score(y_test, y_pred_des)
print(f"Accuracy of DecisionTreeClassifier: {accuracy_des}")
```

```python
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred_des)
print("Precesion of Decesion tree",precision)
```

```python
from sklearn.metrics import recall_score
recall = recall_score(y_test, y_pred_des)
print("recall of Decesion tree",recall)
```

*Visual*

```python
from sklearn.metrics import confusion_matrix
mat_des = confusion_matrix(y_pred_des, y_test)
print("Confusion matrix:")
print(mat_des)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(mat_des, annot=True, fmt='d', cmap='Pastel1_r',
            xticklabels=['Class 0 (Rain)', 'Class 1 (Snow)'],
            yticklabels=['Class 0 (Rain)', 'Class 1 (snow)'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Decesion Tree Heatmap')
plt.show()
```
